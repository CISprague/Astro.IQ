# Sorting a list of initial states
In the process of implementing continuation (homotopy) in generating a data set of optimal control trajectories, it is important that each subsequent initial state is similar to the last.

```python
# First we import a model for example
import sys
sys.path.append('../')
from Trajectory import Point_Lander
from numpy import *
```

```python
# We define arbitrary starting bounds for the state
silb = array([-500, 500, -200, -200, 8000], float)
siub = array([500, 1000, 200, 200, 9500], float)
```

```python
# We generate a list of random intiial states
ntraj = 50
statesi = [] # List of states
rl      = [] # List of distances from target
for i in range(ntraj):
    state = random.random()*(siub - silb) + silb
    statesi.append(state)
    # We create a list of distances by which to sort
    rl.append(linalg.norm(state[0:4]))
# We compute the indicies of the sorted distance list
order = argsort(rl)
statesi = array(statesi)
siorder = statesi[order]
siorder[:,0:4]
```
